# **2D Transformation-Rotation**

```python
# Rotation:

import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
import cv2
import math

img = cv2.imread("C:/Users/HP/Pictures/Saved Pictures/cube 3x3.png",0)
h, w = img.shape

nimg = np.zeros((w+250, h+250), dtype = "uint8")
radians = math.radians(30)

for i in range(h):
    for j in range(w):
        pix = img[i][j]
        pnt = np.array([i,j]).reshape(-1,1)
        rotm = np.array([math.cos(radians), -math.sin(radians), math.sin(radians), math.cos(radians)])
        rotm = rotm.reshape(2,2)
        newmat = rotm @ pnt
        xf = newmat[0][0]
        yf = newmat[1][0]
        x2 = int(xf+150)
        y2 = int(yf)
        nimg[x2,y2] = pix
        
fig, ax = plt.subplots()
ax.imshow(nimg, extent = [0, 350, 200, -150], cmap = 'gray')
plt.show()
```
