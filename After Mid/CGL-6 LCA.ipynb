# Line Clipping Algorithm

```python
x1, y1 = input().split()
x2, y2 = input().split()
```

```python
x1 = int(x1)
y1 = int(y1)
x2 = int(x2)
y2 = int(y2)
```

```python
xmax = int(input("Xmax:"))
xmin = int(input("Xmin:"))
ymax = int(input("Ymax:"))
ymin = int(input("Ymin:"))
```

```python
y = (y2 - y1)
x = (x2 - x1)
m = y/x
print(m)
```

```python
n_x = x1 + ((ymin - y1) / m)
if xmin <= n_x <= xmax:
    print((n_x, ymin))

n_x = x1 + ((ymax - y1) / m)
if xmin <= n_x <= xmax:
    print((n_x, ymax))

n_y = y1 + (m * (xmin - x1))
if ymin <= n_y <= ymax:
    print((xmin, n_y))

n_y = y1 + (m * (xmax - x1))
if ymin <= n_y <= ymax:
    print((xmax, n_y))
```
